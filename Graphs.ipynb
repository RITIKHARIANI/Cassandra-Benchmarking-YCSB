```python
#Workload A
import matplotlib.pyplot as plt

#Data Processing
f = open("Workload_a_rg_values_RUN.txt","r")
data = f.read()
data = data.split('\n')

record_count=[100000,200000,300000,400000,500000]
read_latency = []
update_latency = []
throughput = []

for i in range(5):
    throughput.append(data[i].split(',')[2])
    
for i in range(30,43,3):
    read_latency.append(data[i].split(',')[2])
    update_latency.append(data[i+2].split(',')[2])
f.close()

print(throughput,len(throughput))
print(read_latency,len(read_latency))
print(update_latency,len(update_latency))

read_latency = list(map(float,read_latency))
update_latency = list(map(float,update_latency))
throughput = list(map(float,throughput))

#Plotting
plt.figure(1)
plt.plot(record_count,read_latency,label="Avg Read_Latency",marker="o")
plt.plot(record_count,update_latency,label="Avg Update_Latency",marker="o")
plt.legend()
plt.title("Workload A")
plt.xlabel("RecordCount")
plt.ylabel("Average Latency (us)")
plt.show()
plt.figure(2)
plt.plot(record_count,throughput,label="Throughput",marker="o",color="red")
plt.legend()
plt.title("Workload A")
plt.xlabel("RecordCount")
plt.ylabel("Throughput")
plt.show()
plt.figure(3)
plt.plot(throughput,read_latency,label="Avg Read_Latency",marker="o")
plt.plot(throughput,update_latency,label="Avg Update_Latency",marker="o")
plt.legend()
plt.title("Workload A")
plt.xlabel("Throughput")
plt.ylabel("Average Latency (us)")
plt.show()
```

```python
#Workload F
import matplotlib.pyplot as plt

#Data Processing
f = open("Workload_f_rg_values_RUN.txt","r")
data = f.read()
data = data.split('\n')

record_count=[100000,200000,300000,400000,500000]
read_latency = []
update_latency = []
rmw = []
throughput = []

for i in range(5):
    throughput.append(data[i].split(',')[2])
    
for i in range(5,22,4):
    read_latency.append(data[i].split(',')[2])
    update_latency.append(data[i+3].split(',')[2])
    rmw.append(data[i+1].split(',')[2])
f.close()

print(throughput,len(throughput))
print(read_latency,len(read_latency))
print(update_latency,len(update_latency))
print(rmw,len(rmw))

read_latency = list(map(float,read_latency))
update_latency = list(map(float,update_latency))
rmw = list(map(float,rmw))
throughput = list(map(float,throughput))

#Plotting
plt.figure(1)
plt.plot(record_count,read_latency,label="Avg Read_Latency",marker="o")
plt.plot(record_count,rmw,label="Avg Read_Modify_Write_Latency",marker="o")
plt.plot(record_count,update_latency,label="Avg Update_Latency",marker="o")
plt.legend()
plt.title("Workload F")
plt.xlabel("RecordCount")
plt.ylabel("Average Latency (us)")
plt.show()
plt.figure(2)
plt.plot(record_count,throughput,label="Throughput",marker="o",color="red")
plt.legend()
plt.title("Workload F")
plt.xlabel("RecordCount")
plt.ylabel("Throughput")
plt.show()
plt.figure(3)
plt.plot(throughput,read_latency,label="Avg Read_Latency",marker="o")
plt.plot(throughput,rmw,label="Avg Read_Modify_Write_Latency",marker="o")
plt.plot(throughput,update_latency,label="Avg Update_Latency",marker="o")
plt.legend()
plt.title("Workload F")
plt.xlabel("Throughput")
plt.ylabel("Average Latency (us)")
plt.show()
```
